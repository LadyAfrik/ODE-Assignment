## **Exercise 1**

```python
#Importing the necessary libraries
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Defining the function that returns dy/dt
def model(y,t):
    k = 0.3
    dydt = -k * y
    return dydt

# Defining the initial condition
y0 = 5

# Difininf the time points
t = np.linspace(0,20)

# Solving ODE by calling the defined model function and passing the two parameters
y = odeint(model,y0,t)

# Plotting results
plt.plot(t,y)
plt.xlabel('time')
plt.ylabel('y(t)')
plt.show()
```

## **Exercise 2**

```python
#Importing the necessary libraries
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
#Defining the time scale to be explored numerically.

# Initialization of the necessary parameters
tstart= 0
tstop= 50
increment = 0.1
x0 = [3.14,-0.5]

#Parameters
a=0
b=0.4
c=-0.3
e=0
```

```python
t = np.arange(tstart,tstop+1,increment)

# Defining the function that returns dx/dt
def mydiff(x, t):
  dRdt = a*x[0]+ b*x[1]
  dJdt = c*x[0]+ e*x[1]

  dxdt= [dRdt,dJdt]

  return dxdt
```

```python
# Solving ODE
x = odeint(mydiff, x0, t)
print(x)

#Printing R and J
R = x[:,0]
J = x[:,1]
```

```python
#Plotting the graph
plt.plot(t,R)
plt.plot(t,J)
plt.title('Simulation with 2 variables')
plt.xlabel('t')
plt.ylabel('Emotions')
plt.grid()
plt.axis([0, 50, -5, 5])
plt.legend(["Romeo", "Juliet"])
plt.show()
```

# **Exercise 2 Result Interpretation**

The above plot is a system models for a turbulent relationship where Juliet's negative feelings initially drive Romeo’s positive feelings to decrease. However, over time, their roles alternate, leading to oscillations where one partner's emotions strongly influence the other, but always in opposing directions.

This behavior reflects an emotional “push-pull” dynamic, where one partner’s feelings affect and trigger the opposite reaction in the other.
